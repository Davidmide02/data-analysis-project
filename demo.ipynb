```python
#  load file
# 1. total number of annotators
# 2. total number of females
# 3. total number of males

df = pd.read_csv("langeasy_audio(19th-to-21st-june).csv")
number_of_people = df['userId'].count()
number_of_female = df['userGender'].value_counts().get('female')
number_of_male = df['userGender'].value_counts().get('male')
gender_counter = df['userGender'].value_counts()


print('people',number_of_people)
print('female',number_of_female)
print('male',number_of_male)
print('total',number_of_female+number_of_male)
print('gender_counter',gender_counter)


```

```python
# 4. gender count

gender_count = sns.countplot(x=df['userGender'])
plt.title('Gender count')
plt.show()
```

```python
#  5. gender distribution
plt.figure(figsize=(8,8))
plt.pie(gender_counter.values, labels=gender_counter.index, autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.show()
```
